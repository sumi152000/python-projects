```python
from tkinter import *
window = Tk()
window.title('Calculator')
window.geometry('220x289')
window.configure(bg="Black")

label=Label(window,width=24,height=2,font="Bold")
label.place(x=0,y=0)

equation=" "

def show(value):
    global equation
    equation+=value
    label.config(text=equation)
    
def clear():
    global equation
    equation = " "
    label.config(text=" ")


def calculate():
    global equation
    try:
        result = eval(equation)
        label.config(text=str(result))
        equation = str(result)
    except Exception as e:
        label.config(text="Error")
#row2

clear=Button(window,text='C',width=18,height=2,font="bold",bg="orange",command=clear)
clear.place(x=0,y=42)

division=Button(window,text='/',width=5,height=2,font="bold",command=lambda:show("/"))
division.place(x=165,y=42)

#row3
seven=Button(window,text='7',width=5,height=2,font="bold",command=lambda:show("7"))
seven.place(x=0,y=92)

eight=Button(window,text='8',width=5,height=2,font="bold",command=lambda:show("8"))
eight.place(x=55,y=92)

nine=Button(window,text='9',width=5,height=2,font="bold",command=lambda:show("9"))
nine.place(x=110,y=92)

multiplication=Button(window,text='x',width=5,height=2,font="bold",command=lambda:show("*"))
multiplication.place(x=165,y=92)

#row4

four=Button(window,text="4",width=5,height=2,font="bold",command=lambda:show("4"))
four.place(x=0,y=141)

five=Button(window,text="5",width=5,height=2,font="bold",command=lambda:show("5"))
five.place(x=55,y=141)

six=Button(window,text="6",width=5,height=2,font="bold",command=lambda:show("6"))
six.place(x=110,y=141)

subtraction=Button(window,text="-",width=5,height=2,font="bold",command=lambda:show("-"))
subtraction.place(x=165,y=141)

#row5
one=Button(window,text="1",width=5,height=2,font="bold",command=lambda:show("1"))
one.place(x=0,y=190)

two=Button(window,text="2",width=5,height=2,font="bold",command=lambda:show("2"))
two.place(x=55,y=190)

three=Button(window,text="3",width=5,height=2,font="bold",command=lambda:show("3"))
three.place(x=110,y=190)

addition=Button(window,text="+",width=5,height=2,font="bold",command=lambda:show("+"))
addition.place(x=165,y=190)

zero=Button(window,text="0",width=5,height=2,font="bold",command=lambda:show("0"))
zero.place(x=0,y=239)

decimal=Button(window,text=".",width=5,height=2,font="bold",command=lambda:show("."))
decimal.place(x=55,y=239)

equals=Button(window,text="=",width=11,height=2,font="bold",command=calculate)
equals.place(x=110,y=239)

window.mainloop()
```
