```python
import random
```

```python
def player_batting():
    player_total=0
    player_run = int(input("player,Enter the number 1-6: "))
    comp_run = random.randint(1,6)
    print(f'player run : {player_run} and computer bowls: {comp_run}')
    player_total+=player_run
    if player_run == comp_run:
        pass
    while player_run != comp_run:
        player_run = int(input("player,Enter the number 1-6: "))
        comp_run = random.randint(1,6)
        player_total+=player_run
        print(f'player enter {player_run} and computer enter {comp_run}')
        print(f'player run : {player_total}')
    return player_total
```

```python
def comp_batting():
    comp_total=0
    player_run = int(input("player,Enter the number 1-6 : "))
    comp_run = random.randint(1,6)
    print(f'computer run : {comp_run} and player bowls : {player_run}')
    comp_total+=comp_run
    if comp_run == player_run:
        pass
    while player_run != comp_run:
        player_run = int(input("player,Enter the number 1-6: "))
        comp_run = random.randint(1,6)
        comp_total+=comp_run
        print(f'player enter {player_run} and computer enter {comp_run}')
        print(f'comp run : {comp_total}')
    return comp_total
```

```python
while True:
    player = input('Player, choose "odd" or "even" : ')
    if player == 'even':
        computer = 'odd'
    else:
        computer = 'even'
            
    print(f" player choose : '{player}' and computer choose : '{computer}'")
    
    player_random = int(input("enter a random number: "))
    comp_random = random.randint(1,6)
    a=player_random+comp_random
    print(f'player : {player_random} and computer : {comp_random}')
    
    if a% 2 == 0:
        batting ="even"
    else:
        batting ="odd"
    print(f"total {a},{batting} battings!")
    
    print("--------------")
    if player =='even' and batting =='even':
        result=player_batting()
        print("out")
        print(f'player total score is {result}')
        result2=comp_batting()
        print("out")
        print(f'computer total score is {result2}')
    elif player=='even' and batting =='odd':
        result2=comp_batting()
        print("out")
        print(f'computer total score is {result2}')
        result=player_batting()
        print("out")
        print(f'player total score is {result}')
    elif player == 'odd' and batting =='odd':
        result=player_batting()
        print("out")
        print(f'player total score is {result}')
        result2=comp_batting()
        print("out")
        print(f'computer total score is {result2}')
    elif player =='odd' and batting =='even':
        result2=comp_batting()
        print("out")
        print(f'computer total score is {result2}')
        result=player_batting()
        print("out")
        print(f'player total score is {result}')
              
    if result> result2:
        print("Player Wins!")
    elif result2> result:
        print("Computer Wins!")
    else:
        print("Tie!")
    break
```
